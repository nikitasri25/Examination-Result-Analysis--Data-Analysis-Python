```python
df= pd.read_csv("Expanded_data_with_more_features.csv")
print(df.head())
```

#drop unnamed column

```python
df= df.drop('Unnamed: 0', axis=1)
```

```python
df.head()
```

#change weekly studying hours

```python
df["WklyStudyHours"]= df["WklyStudyHours"].str.replace("05-Oct", "5-10")
df.head()
```

#gender distribution

```python
plt.figure(figsize=(5,5))
ax=sns.countplot(data=df, x= "Gender")
plt.title("Gender Distribution")
ax.bar_label(ax.containers[0])
plt.show()
```

From the above chart it is analysed that number of females are more than the number of males

```python
gb= df.groupby("ParentEduc").agg({"MathScore": 'mean', "ReadingScore": 'mean', "WritingScore":'mean'})
print(gb)
```

```python
sns.heatmap(gb, annot=True)
plt.title("Parents' Education Vs Students'Scores")
plt.show()
```

From the above graph it can be analyzed that parents' education play an important role in students' scores

```python
gb1= df.groupby("ParentMaritalStatus").agg({"MathScore": 'mean', "ReadingScore": 'mean', "WritingScore":'mean'})
print(gb1)
```

```python
sns.heatmap(gb1, annot=True)
plt.title("Parents' Marital Status Vs Students' Scores")
plt.show()
```

It can be concluded that there is no significant impact on students' scores based on their parents' marital status

Distribution of ethnic groups

```python
groupA_count = df[df['EthnicGroup'] == "group A"]['EthnicGroup'].count()
groupB_count = df[df['EthnicGroup'] == "group B"]['EthnicGroup'].count()
groupC_count = df[df['EthnicGroup'] == "group C"]['EthnicGroup'].count()
groupD_count = df[df['EthnicGroup'] == "group D"]['EthnicGroup'].count()
groupE_count = df[df['EthnicGroup'] == "group E"]['EthnicGroup'].count()

mlist = [groupA_count, groupB_count, groupC_count, groupD_count, groupE_count]
l = ["group A", "group B", "group C", "group D", "group E"]
plt.title("Distribution of Ethnic Groups")
plt.pie(mlist, labels=l, autopct="%1.2f%%")
plt.show()
```
